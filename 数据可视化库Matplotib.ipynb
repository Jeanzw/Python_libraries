```python
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib inline  #多加了这一行其实就不需要用plt.show()就可以每一次plt.plot()的时候图像会自动显示，不过这个在thonny里面好像用不了，在Jupiter里面可以用
x = np.linspace(0,2*np.pi,100)  #通过np.linspace实现，从0到100，间隔2个pi，生成一个series x
y = np.sin(x)  #通过sin函数把x当做参数传进去
plt.plot(x,y)
plt.show()
```

## 2. matplotlib简单绘图之plot

### （4）视图中两条线并个性化视图

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure() 
t = np.arange(0.0,2.0,0.1)
s = np.sin(t*np.pi)
print(s)
plt.plot(t,s,'r--', label = 'aaaa')  #在同一个图片里面展现多条线，我这里相当于一条画的是以t为x轴，s为y轴，然后红色虚线的sin函数。另一条相当于一条画的是以2t为x轴，s为y轴，然后蓝色虚线的sin函数
plt.plot(t*2, s,'--',label = 'bbbb')
plt.xlabel('This is x')  #给x和y轴上label
plt.ylabel('This is y')
plt.title('This is a demo') #给图片上标题
plt.legend()  #给视图上图例
plt.show()
```

## 3. matplotlib简单绘图之subplot

### （1）上节课的绘图

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure() 
x = np.linspace(0.0,5.0)
y1 = np.sin(np.pi * x)
y2 = np.sin(np.pi * x * 2)

plt.plot(x, y1,'b--', label = 'sin(pi * x)')
plt.ylabel('This is y1')
plt.plot(x, y2, 'r--', label = 'sin(np.pi * 2x)')
plt.ylabel('This is y2')
plt.xlabel('This is x')  
plt.title('This is a x-y value') 
plt.legend() 
plt.show()
```

### （2）画子图
上面的图就是上节课我们学到的内容展现的图，那么我们想要把这张图通过subplot展示

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure() 
x = np.linspace(0.0,5.0)
y1 = np.sin(np.pi * x)
y2 = np.sin(np.pi * x * 2)

plt.subplot(2,1,1)   #我们要生成2行1列的子图，他的序列是1
plt.plot(x, y1, 'b--')
plt.ylabel('y1')
plt.subplot(2,1,2)
plt.plot(x, y2, 'r--')   #我们要生成2行1列的子图，他的序列是2
plt.ylabel('y2')
plt.xlabel('This is x')
plt.show()   #这里相当于我们把之前的图里面的两条线分别拿出来单独成图，两行一列就相当于是两个子图上下排列
```

## 4. Pandas绘图之Series

### （1）Series绘图初探

```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt

#创建一个series
s1 = Series(np.random.randn(100)).cumsum()  #随机创建1000个数，然后对其求累加和。
#所谓累加和
#s = Series([1,2,3,4,5])
#print(s)
#print(s.cumsum())
s1.plot()   #直接对s1进行绘图，并没有传入任何参数
plt.show()
```

### （2）Series绘图多参数

```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt

s1 = Series(np.random.randn(100)).cumsum()  
#第一个参数，kind,默认line，可以是bar, pie
#第二个参数是grid，即是否有格子
#第三个参数是label,图例
#第四个参数title
#第五个参数style
s1.plot(kind = 'line',grid = True, label = 'S1', title = 'This is Series', style = '--')
plt.legend()
plt.show()
```

### （3）在一张图里画两个series的数据

```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt

s1 = Series(np.random.randn(100)).cumsum()
s2 = Series(np.random.randn(100)).cumsum()
s1.plot(kind = 'line',grid = True, label = 'S1', title = 'This is Series', style = '--')
s2.plot(label = 'S2')
plt.legend()
plt.show()
```

### （4）通过subplot去画两个series数据

```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt

s1 = Series(np.random.randn(100)).cumsum()
s2 = Series(np.random.randn(100)).cumsum()

fig, ax = plt.subplots(2,1)  #生成2行1列的子图
#ax其实代表的就是子图，有两个元素的array,[0]代表的是第一张子图，[1]代表的是第二张子图
ax[0].plot(s1)  #在第一张子图上plots1
ax[1].plot(s2)
plt.show()
```

```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt

s1 = Series(np.random.randn(100)).cumsum()
s2 = Series(np.random.randn(100)).cumsum()
#除此之外，我们还可以对每个子图进行调整
fig, ax = plt.subplots(2,1)
s1.plot(ax = ax[0],label = 'S1',kind = 'bar')
s2.plot(ax = ax[1],label = 'S2')
plt.legend
plt.show()
```

## 6. 直方图和密度图：了解数据分布的情况

### （1）直方图初探

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt

s = np.arange(10)
print(s)
plt.hist(s, rwidth = 0.9)  #设置了一个宽度0.9,也就是每一个柱子的间隔
plt.show()  #画的其实就是s的值的分布，比如说值为0的，那么只有一个，值为1的也就只有一个
```

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
#加大数据
s1 = Series(np.random.randn(10))
print(s1)
plt.hist(s1, rwidth = 0.9)
plt.show()  #也就相当于这10个数据的分布情况，但是要注意，这里的横轴是一个区间，也就是相当于说有多少个值是分布在这个区间之内的
```

### （2）直方图的参数

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
#刚刚已经有一个rwidth
#第二个参数，bins:分隔区间，默认为10,可以调整
#第三个参数，color
#第四个参数，orientation，默认vertical可以改成horizontal
plt.hist(s1, rwidth = 0.9, bins = 20, color = 'r', orientation = 'horizontal')
plt.show()
```

### （3）密度图

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
s = Series(np.random.randn(100))
s.plot(kind = 'kde')
plt.show()  #这个时候显示出来的就是一个密度图
```

## 7. 基本图形绘制

### （1）line chart:通常是时间的统计情况

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
x = np.random.randn(5)
y = np.random.randn(5)
plt.plot(x,y)
plt.show()
```

### （2）dual line

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
x = [1,2,3,4,5]
y1 = np.random.randn(5)
y2 = np.random.randn(5)
plt.plot(x, y1)
plt.plot(x, y2)
#add label
plt.plot(x, y1, label = 'line1')
plt.plot(x, y2, label = 'line2')
plt.legend()
#add title
plt.title('sample line chart')
plt.ylabel('ratio')
plt.xlabel('period')
plt.show()
```

### （3）bar chart:表示分类的统计情况

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
x = [1,2,3,4,5]
y3 = [0.3, 0.4, -0.7, -0.5,0.6]
plt.bar(x, y3,label='region 1')
plt.legend()
plt.title('simple bar chart')
plt.ylabel('ratio')
plt.xlabel('period')
plt.show()
```

### （4）Stacked bar chart

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
x = [1,2,3,4,5]
y3 = [0.3, 0.4, -0.7, -0.5,0.6]
y4 = [0.2,0.4,0.7,-0.9,-0.4]
plt.bar(x, y3,label='region 1')
plt.bar(x, y4,label='region 2')
plt.legend()
plt.title('simple bar chart')
plt.ylabel('ratio')
plt.xlabel('period')
plt.show()
```

### （5）area chart:通常会表示面积变化或者百分比变化状况

#### 无标签

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
x = [1,2,3,4,5]
y3 = [0.3, 0.4, -0.7, -0.5,0.6]
y4 = [0.2,0.4,0.7,-0.9,-0.4]
y5 = [0.5,0.4,0.2,-0.8,-0.6]
plt.stackplot(x, y3,y4,y5, colors=['blue','orange','red'])
plt.title('simple area chart')
plt.ylabel('ratio')
plt.xlabel('period')
plt.legend()
plt.show()
```

#### 有标签

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
x = [1,2,3,4,5]
y3 = [0.3, 0.4, -0.7, -0.5,0.6]
y4 = [0.2,0.4,0.7,-0.9,-0.4]
y5 = [0.5,0.4,0.2,-0.8,-0.6]

plt.plot([],[],label = 'region 1',color = 'blue') #注意这里是color而不是colors
plt.plot([],[],label = 'region 2',color = 'orange')#如果不写color则和表中area chart对应颜色不一样
plt.plot([],[],label = 'region 3',color = 'red')

plt.stackplot(x, y3,y4,y5, colors=['blue','orange','red'])
plt.legend()
plt.title('simple area chart')
plt.ylabel('ratio')
plt.xlabel('period')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
size = [2,3,4,5,10]
tag = ['region 1','region 2','region 3','region 4', 'region 5']
plt.pie(size,startangle = 90,shadow = True, labels = tag) #加了阴影之后看上去比较立体

plt.show()
```

#### 强调作用

```python
#我要把某一部分抽出来，起到强调作用
import matplotlib.pyplot as plt
import numpy as np
plt.figure()
size = [2,3,4,5,10]
tag = ['region 1','region 2','region 3','region 4', 'region 5']
plt.pie(size,startangle = 90,shadow = True,explode = [0.1,0.09,0.08,0.07,0.06], labels = tag) 
#explode = []其实就是弹出的幅度
```
